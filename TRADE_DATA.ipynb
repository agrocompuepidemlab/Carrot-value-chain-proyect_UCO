```python
# Save this as plot_nutricosmetics.py and run in the same environment that has access to the given path.
import os
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
import textwrap

plt.rcParams.update({
    "font.family": "serif",          # Times-like for article style
    "font.serif": ["Times New Roman", "Times", "DejaVu Serif"],
    "figure.dpi": 300,
    "axes.titlesize": 12,
    "axes.labelsize": 10,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
    "legend.fontsize": 9
})

# ---------- USER CONFIG ----------
EXCEL_PATH = "/content/drive/MyDrive/Proyecto Zanahoria/Data_Agrosavia/Estudio de percepcion y perfil del consumidor de Nutricosmeticos - Actualizado1.xlsx"
SHEET_NAME = "Sheet1"
OUTPUT_DIR = "./figures_nutricosmetics"
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)
# ---------------------------------

# Column names (as provided). If your file has slightly different names, update these strings.
col_heard_points = "Puntos: Ha odo hablar de Nutricosmticos? \"Los Nutricosmticos son complementos alimenticios que buscan la belleza exterior desde el interior. Al ingerir los nutrientes por va oral se consigue que los ...\""
col_heard_text = "Ha odo hablar de Nutricosmticos? \"Los Nutricosmticos son complementos alimenticios que buscan la belleza exterior desde el interior. Al ingerir los nutrientes por va oral se consigue que los ...\""

col_used_points = "Puntos: Ha usado o consumido Nutricosmticos? "
col_used_text = "Ha usado o consumido Nutricosmticos? "

col_motives = "Qu motivos lo han llevado a comprar este tipo de productos?"
col_motives_points = "Puntos: Qu motivos lo han llevado a comprar este tipo de productos?"

col_age = "Rango de edad en el que se encuentra"
col_age_points = "Puntos: Rango de edad en el que se encuentra"

col_sex = "Sexo"
col_sex_points = "Puntos: Sexo"

col_products_used = " Qu productos de Nutricosmticos utiliza?"
col_products_used_points = "Puntos:  Qu productos de Nutricosmticos utiliza?"

# Helper: load sheet
if not os.path.isfile(EXCEL_PATH):
    raise FileNotFoundError(f"Excel file not found at: {EXCEL_PATH}\nIf running in Colab, first mount drive with:\nfrom google.colab import drive\ndrive.mount('/content/drive')")

df = pd.read_excel(EXCEL_PATH, sheet_name=SHEET_NAME)
print("Loaded dataframe with shape:", df.shape)

# Helper: try to find existing column among alternatives
def find_col(preferred, alternatives=[]):
    if preferred in df.columns:
        return preferred
    for alt in alternatives:
        if alt in df.columns:
            return alt
    # try fuzzy by lowercase contains
    lower_cols = {c.lower(): c for c in df.columns}
    key = preferred.lower()
    for lc, orig in lower_cols.items():
        if key in lc:
            return orig
    return None

# Attempt to resolve columns (robustness)
col_heard_points = find_col(col_heard_points, [col_heard_points, "Puntos: Ha odo hablar de Nutricosmticos?", col_heard_text])
col_used_points = find_col(col_used_points, [col_used_points, "Puntos: Ha usado o consumido Nutricosmticos?"])
col_motives = find_col(col_motives, [col_motives, "Qu motivos lo han llevado a comprar este tipo de productos?"])
col_age = find_col(col_age, [col_age, "Rango de edad en el que se encuentra"])
col_sex = find_col(col_sex, [col_sex, "Sexo"])
col_products_used = find_col(col_products_used, [col_products_used])

print("Columns used (resolved):")
print("heard points:", col_heard_points)
print("used points: ", col_used_points)
print("motives:     ", col_motives)
print("age:         ", col_age)
print("sex:         ", col_sex)
print("products:    ", col_products_used)

# Generic cleaner for multi-select answers (comma / ; / /)
def split_multianswer(cell):
    if pd.isna(cell):
        return []
    if isinstance(cell, (int, float)):
        return [str(cell)]
    text = str(cell)
    # common separators
    parts = []
    for sep in [";", ",", "/", "|"]:
        if sep in text:
            parts = [p.strip() for p in text.split(sep) if p.strip()]
            break
    if not parts:
        # also split by multiple spaces if looks like a list
        parts = [text.strip()] if text.strip() else []
    return parts

# 1) Heard vs Used (proportion plot)
def plot_heard_vs_used(df):
    # Use points columns if present (assume >0 means yes), otherwise text columns
    heard_col = col_heard_points or col_heard_text
    used_col = col_used_points or col_used_text
    if heard_col is None or used_col is None:
        print("Missing columns for heard/used analysis, skipping.")
        return
    # convert to binary yes/no
    def to_yesno(val):
        if pd.isna(val):
            return "No response"
        if isinstance(val, (int, float)):
            # assume 1..5 scale possible: treat >0 as yes
            return "Yes" if val > 0 else "No"
        s = str(val).strip().lower()
        if s in ["yes", "y", "si", "s", "s", "1"]:
            return "Yes"
        if s in ["no", "n", "0", "no response", "nan", ""]:
            return "No"
        # if longer descriptive: try to detect 'no' word
        if "no" in s and len(s) < 10:
            return "No"
        # default: if text present consider yes
        return "Yes"
    heard = df[heard_col].map(to_yesno)
    used = df[used_col].map(to_yesno)

    # cross-tab
    ctab = pd.crosstab(heard, used, normalize='index')  # proportions by heard row
    # plot
    fig, ax = plt.subplots(figsize=(5,4))
    # stacked bar: for each heard category, show distribution of used
    categories = list(ctab.columns)
    bottoms = np.zeros(len(ctab))
    x = np.arange(len(ctab))
    for col in categories:
        vals = ctab[col].values
        ax.bar(x, vals, bottom=bottoms, label=col)
        bottoms += vals
    ax.set_xticks(x)
    ax.set_xticklabels([str(i) for i in ctab.index], rotation=0)
    ax.set_ylim(0,1)
    ax.set_ylabel("Proportion")
    ax.set_xlabel("Heard about nutricosmetics?")
    ax.set_title("Heard vs Used (proportion)")
    ax.legend(title="Used", loc="upper right", frameon=False)
    fig.tight_layout()
    out = os.path.join(OUTPUT_DIR, "heard_vs_used.png")
    fig.savefig(out, dpi=300)
    print("Saved:", out)

# 2) Relation of purchase motives with generation (age range)
def plot_motives_by_age(df, top_n=8):
    if col_motives is None or col_age is None:
        print("Missing columns for motives-by-age, skipping.")
        return
    # explode multi-select motives
    rows = []
    for _, r in df[[col_age, col_motives]].iterrows():
        age = r[col_age] if not pd.isna(r[col_age]) else "Unknown"
        parts = split_multianswer(r[col_motives])
        if not parts:
            rows.append((age, "No answer"))
        else:
            for p in parts:
                rows.append((age, p))
    temp = pd.DataFrame(rows, columns=["age", "motive"])
    # compute top motives overall
    top = [m for m,c in Counter(temp['motive']).most_common(top_n)]
    temp = temp[temp['motive'].isin(top)]
    # pivot: age x motive counts (normalized per age)
    pivot = pd.crosstab(temp['age'], temp['motive'])
    pivot_norm = pivot.div(pivot.sum(axis=1).replace(0,1), axis=0)  # proportions per age
    # sort ages by natural order if possible (attempt numeric)
    try:
        # try to order by min age in range (e.g. '18-24')
        def age_key(a):
            import re
            m = re.search(r'(\d{2,4})', str(a))
            return int(m.group(1)) if m else 9999
        pivot_norm = pivot_norm.reindex(sorted(pivot_norm.index, key=age_key))
    except Exception:
        pass

    fig, ax = plt.subplots(figsize=(8,4.5))
    bottom = np.zeros(pivot_norm.shape[0])
    x = np.arange(pivot_norm.shape[0])
    for col in pivot_norm.columns:
        vals = pivot_norm[col].values
        ax.bar(x, vals, bottom=bottom, label=textwrap.fill(col, 25))
        bottom += vals
    ax.set_xticks(x)
    ax.set_xticklabels(pivot_norm.index, rotation=45, ha='right')
    ax.set_ylabel("Proportion within age range")
    ax.set_xlabel("Age range")
    ax.set_title("Top purchase motives by age range")
    ax.legend(title="Motive", bbox_to_anchor=(1.02, 1), loc='upper left', frameon=False)
    fig.tight_layout()
    out = os.path.join(OUTPUT_DIR, "motives_by_age.png")
    fig.savefig(out, dpi=300)
    print("Saved:", out)

# 3) Relation of purchase motive according to sex
def plot_motives_by_sex(df, top_n=8):
    if col_motives is None or col_sex is None:
        print("Missing columns for motives-by-sex, skipping.")
        return
    rows = []
    for _, r in df[[col_sex, col_motives]].iterrows():
        sex = r[col_sex] if not pd.isna(r[col_sex]) else "Unknown"
        parts = split_multianswer(r[col_motives])
        if not parts:
            rows.append((sex, "No answer"))
        else:
            for p in parts:
                rows.append((sex, p))
    temp = pd.DataFrame(rows, columns=["sex", "motive"])
    top = [m for m,c in Counter(temp['motive']).most_common(top_n)]
    temp = temp[temp['motive'].isin(top)]
    pivot = pd.crosstab(temp['sex'], temp['motive'])
    pivot_norm = pivot.div(pivot.sum(axis=1).replace(0,1), axis=0)
    fig, ax = plt.subplots(figsize=(6,4))
    bottom = np.zeros(pivot_norm.shape[0])
    x = np.arange(pivot_norm.shape[0])
    for col in pivot_norm.columns:
        vals = pivot_norm[col].values
        ax.bar(x, vals, bottom=bottom, label=textwrap.fill(col,25))
        bottom += vals
    ax.set_xticks(x)
    ax.set_xticklabels(pivot_norm.index, rotation=0)
    ax.set_ylabel("Proportion within sex")
    ax.set_xlabel("Sex")
    ax.set_title("Top purchase motives by sex")
    ax.legend(title="Motive", bbox_to_anchor=(1.02,1), loc='upper left', frameon=False)
    fig.tight_layout()
    out = os.path.join(OUTPUT_DIR, "motives_by_sex.png")
    fig.savefig(out, dpi=300)
    print("Saved:", out)

# 4) Nutricosmetics used (top products)
def plot_products_used(df, top_n=15):
    if col_products_used is None:
        print("Missing column for products used, skipping.")
        return
    all_items = []
    for val in df[col_products_used].dropna().astype(str):
        parts = split_multianswer(val)
        all_items.extend(parts)
    counts = Counter([it.strip() for it in all_items if it.strip()])
    if not counts:
        print("No product responses found.")
        return
    top = counts.most_common(top_n)
    labels, values = zip(*top)
    fig, ax = plt.subplots(figsize=(6, max(3, 0.35*len(labels))))
    y = np.arange(len(labels))
    ax.barh(y, values)
    ax.set_yticks(y)
    ax.set_yticklabels(labels)
    ax.invert_yaxis()
    ax.set_xlabel("Count")
    ax.set_title("Most used nutricosmetics (top {})".format(len(labels)))
    fig.tight_layout()
    out = os.path.join(OUTPUT_DIR, "products_used_top.png")
    fig.savefig(out, dpi=300)
    print("Saved:", out)

# Run plots
plot_heard_vs_used(df)
plot_motives_by_age(df, top_n=8)
plot_motives_by_sex(df, top_n=8)
plot_products_used(df, top_n=15)

print("All done. Figures saved in:", OUTPUT_DIR)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Primero, cargar los datos e identificar los nombres de columnas
file_path = '/content/drive/MyDrive/Proyecto Zanahoria/Data_Agrosavia/Estudio de percepcion y perfil del consumidor de Nutricosmeticos - Actualizado1.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Identificar nombres exactos de columnas
print("Nombres de columnas disponibles:")
for i, col in enumerate(df.columns):
    print(f"{i}: {col}")

# Buscar columnas por palabras clave
target_keywords = {
    'heard_about': ['odo hablar', 'nutricosmticos'],
    'used': ['usado', 'consumido'],
    'purchase_reasons': ['motivos', 'comprar'],
    'age_group': ['edad', 'rango'],
    'gender': ['sexo'],
    'products_used': ['productos', 'utiliza']
}

available_columns = {}
for key, keywords in target_keywords.items():
    for col in df.columns:
        if any(keyword.lower() in col.lower() for keyword in keywords):
            available_columns[key] = col
            print(f"Encontrado: {key} -> {col}")
            break

# Verificar que tenemos todas las columnas necesarias
required_columns = ['heard_about', 'used', 'purchase_reasons', 'age_group', 'gender', 'products_used']
missing_columns = [col for col in required_columns if col not in available_columns]

if missing_columns:
    print(f"Columnas faltantes: {missing_columns}")
    for col in missing_columns:
        print(f"Buscando alternativa para: {col}")
        for actual_col in df.columns:
            if col in actual_col.lower() or any(word in actual_col.lower() for word in col.split('_')):
                available_columns[col] = actual_col
                print(f"Alternativa encontrada: {col} -> {actual_col}")
                break

#  Ajuste manual si no detecta "products_used"
if 'products_used' not in available_columns:
    available_columns['products_used'] = ' Qu productos de Nutricosmticos utiliza?'

print(f"\nColumnas mapeadas: {available_columns}")

# Configuracin de estilo
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 10
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

# Usar los nombres exactos de columnas encontrados
heard_col = available_columns['heard_about']
used_col = available_columns['used']
reasons_col = available_columns['purchase_reasons']
age_col = available_columns['age_group']
gender_col = available_columns['gender']
products_col = available_columns['products_used']

# 1. Grfico: Ha odo hablar vs Ha usado nutricosmticos
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Consumer Perception and Usage of Nutricosmetics', fontsize=16, fontweight='bold')

# 1a. Conocimiento vs Uso
heard_about = df[heard_col].value_counts()
used = df[used_col].value_counts()

categories = ['Yes', 'No']
heard_counts = [
    heard_about.get('S', 0) + heard_about.get('Si', 0) + heard_about.get('Yes', 0),
    heard_about.get('No', 0) + heard_about.get('Not', 0)
]
used_counts = [
    used.get('S', 0) + used.get('Si', 0) + used.get('Yes', 0),
    used.get('No', 0) + used.get('Not', 0)
]

x = np.arange(len(categories))
width = 0.35

bars1 = ax1.bar(x - width/2, heard_counts, width, label='Heard About', alpha=0.8)
bars2 = ax1.bar(x + width/2, used_counts, width, label='Used', alpha=0.8)

ax1.set_xlabel('Response')
ax1.set_ylabel('Count')
ax1.set_title('Awareness vs. Usage of Nutricosmetics')
ax1.set_xticks(x)
ax1.set_xticklabels(categories)
ax1.legend()
ax1.grid(True, alpha=0.3)

for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom')

for bar in bars2:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom')

# 2. Relacin de motivos de compra con la generacin
purchase_reasons = df[reasons_col].dropna()
age_groups = df[age_col].dropna()

min_length = min(len(purchase_reasons), len(age_groups))
reason_age_df = pd.DataFrame({
    'Purchase_Reason': purchase_reasons.iloc[:min_length],
    'Age_Group': age_groups.iloc[:min_length]
})

reason_age_count = reason_age_df.groupby(['Age_Group', 'Purchase_Reason']).size().unstack(fill_value=0)
top_reasons = reason_age_df['Purchase_Reason'].value_counts().head(5).index
reason_age_filtered = reason_age_df[reason_age_df['Purchase_Reason'].isin(top_reasons)]
reason_age_count_filtered = reason_age_filtered.groupby(['Age_Group', 'Purchase_Reason']).size().unstack(fill_value=0)

reason_age_count_filtered.plot(kind='bar', ax=ax2, width=0.8)
ax2.set_xlabel('Age Group')
ax2.set_ylabel('Count')
ax2.set_title('Top 5 Purchase Reasons by Age Group')
ax2.legend(title='Purchase Reasons', bbox_to_anchor=(1.05, 1), loc='upper left')
ax2.grid(True, alpha=0.3)
ax2.tick_params(axis='x', rotation=45)

# 3. Relacin motivo de compra segn el sexo
gender_data = df[gender_col].dropna()
min_length = min(len(purchase_reasons), len(gender_data))
reason_gender_df = pd.DataFrame({
    'Purchase_Reason': purchase_reasons.iloc[:min_length],
    'Gender': gender_data.iloc[:min_length]
})

reason_gender_filtered = reason_gender_df[reason_gender_df['Purchase_Reason'].isin(top_reasons)]
reason_gender_count_filtered = reason_gender_filtered.groupby(['Gender', 'Purchase_Reason']).size().unstack(fill_value=0)

reason_gender_count_filtered.plot(kind='bar', ax=ax3, width=0.8)
ax3.set_xlabel('Gender')
ax3.set_ylabel('Count')
ax3.set_title('Top 5 Purchase Reasons by Gender')
ax3.legend(title='Purchase Reasons', bbox_to_anchor=(1.05, 1), loc='upper left')
ax3.grid(True, alpha=0.3)
ax3.tick_params(axis='x', rotation=45)

# 4. Nutricosmticos usados (top 10)
products_used = df[products_col].dropna()
all_products = []
for products in products_used:
    if pd.notna(products):
        product_list = [p.strip() for p in str(products).split(',')]
        all_products.extend(product_list)

product_counts = pd.Series(all_products).value_counts().head(10)

product_counts.plot(kind='barh', ax=ax4, color='skyblue')
ax4.set_xlabel('Count')
ax4.set_ylabel('Product Type')
ax4.set_title('Top 10 Most Used Nutricosmetic Products')
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Grficos adicionales
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

cross_tab = pd.crosstab(df[heard_col], df[used_col])
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='YlOrRd', ax=axes[0])
axes[0].set_title('Correlation: Awareness vs. Usage')
axes[0].set_xlabel('Has Used')
axes[0].set_ylabel('Has Heard About')

age_distribution = df[age_col].value_counts()
age_distribution.plot(kind='pie', autopct='%1.1f%%', ax=axes[1], startangle=90)
axes[1].set_title('Age Group Distribution')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()

# Estadsticas descriptivas
print("="*50)
print("DESCRIPTIVE STATISTICS")
print("="*50)
print(f"Total respondents: {len(df)}")
print(f"Have heard about nutricosmetics: {heard_counts[0]} ({heard_counts[0]/len(df)*100:.1f}%)")
print(f"Have used nutricosmetics: {used_counts[0]} ({used_counts[0]/len(df)*100:.1f}%)")
print(f"Most common age group: {age_distribution.idxmax()}")
if not purchase_reasons.empty:
    print(f"Most common purchase reason: {purchase_reasons.value_counts().idxmax()}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib import rcParams
import matplotlib.colors as mcolors

# Configurar estilo con paleta Oranges
plt.style.use('default')
sns.set_style("whitegrid")

# Configurar fuente para publicaciones cientficas
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans']
rcParams['font.size'] = 12
rcParams['axes.labelsize'] = 14
rcParams['axes.titlesize'] = 16
rcParams['xtick.labelsize'] = 12
rcParams['ytick.labelsize'] = 12
rcParams['legend.fontsize'] = 12

# Usar la paleta Oranges de matplotlib
oranges_palette = plt.cm.Oranges(np.linspace(0.3, 0.9, 10))
oranges_palette = [mcolors.to_hex(color) for color in oranges_palette]

# Cargar y preparar datos
file_path = '/content/drive/MyDrive/Proyecto Zanahoria/Data_Agrosavia/Estudio de percepcion y perfil del consumidor de Nutricosmeticos - Actualizado1.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# [Tu cdigo de mapeo de columnas aqu]
# ...

# Funcin para traducir respuestas comunes del espaol al ingls
def translate_responses(series):
    translated = series.copy()

    # Mapeo de traduccin
    translation_map = {
        # S/No
        'S': 'Yes', 'Si': 'Yes', 'S ': 'Yes', 'Si ': 'Yes',
        'No': 'No', 'No ': 'No',

        # Gnero
        'Femenino': 'Female', 'Femenino ': 'Female', 'FEMENINO': 'Female',
        'Masculino': 'Male', 'Masculino ': 'Male', 'MASCULINO': 'Male',
        'Otro': 'Other', 'Otro ': 'Other',

        # Rangos de edad - traducir al ingls
        '18-25': '18-25', '26-35': '26-35', '36-45': '36-45',
        '46-55': '46-55', '56-65': '56-65', '65+': '65+',
        'Menor de 18': 'Under 18', 'Mayor de 65': '65+',

        # Razones de compra comunes
        'Calidad': 'Quality', 'Precio': 'Price', 'Recomendacin': 'Recommendation',
        'Eficacia': 'Effectiveness', 'Disponibilidad': 'Availability',
        'Marca': 'Brand', 'Ingredientes naturales': 'Natural ingredients',
        'Salud': 'Health', 'Belleza': 'Beauty', 'Bienestar': 'Wellness',
        'Resultados': 'Results', 'Innovacin': 'Innovation'
    }

    # Aplicar traduccin
    if translated.dtype == 'object':
        translated = translated.map(lambda x: translation_map.get(str(x).strip(), x))

    return translated

# Aplicar traducciones a las columnas relevantes
df[heard_col] = translate_responses(df[heard_col])
df[used_col] = translate_responses(df[used_col])
df[gender_col] = translate_responses(df[gender_col])
df[reasons_col] = translate_responses(df[reasons_col])
df[age_col] = translate_responses(df[age_col])

# 1. GRFICA: Conocimiento vs Uso
plt.figure(figsize=(10, 7))
heard_about = df[heard_col].value_counts()
used = df[used_col].value_counts()

categories = ['Yes', 'No']
heard_counts = [heard_about.get('Yes', 0), heard_about.get('No', 0)]
used_counts = [used.get('Yes', 0), used.get('No', 0)]

x = np.arange(len(categories))
width = 0.35

bars1 = plt.bar(x - width/2, heard_counts, width, label='Heard About',
                color=oranges_palette[2], alpha=0.9, edgecolor='white', linewidth=1.5)
bars2 = plt.bar(x + width/2, used_counts, width, label='Used',
                color=oranges_palette[6], alpha=0.9, edgecolor='white', linewidth=1.5)

plt.xlabel('Response', fontweight='bold')
plt.ylabel('Number of Respondents', fontweight='bold')
plt.title('Awareness vs. Usage of Nutricosmetics', fontsize=16, fontweight='bold')
plt.xticks(x, categories)
plt.legend(frameon=True, fancybox=True, shadow=True, loc='upper right')
plt.grid(True, alpha=0.3)

# Aadir valores en las barras
for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom', fontsize=11, fontweight='bold')

for bar in bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

# 2. GRFICA ESPECFICA: Nutricosmticos usados (vitaminas, omega, etc.)
plt.figure(figsize=(12, 8))
products_used = df[products_col].dropna()
all_products = []
for products in products_used:
    if pd.notna(products):
        product_list = [p.strip() for p in str(products).split(',')]
        # Traduccin y categorizacin especfica de nutricosmticos
        translated_products = []
        for product in product_list:
            product_lower = product.lower()
            if 'colgeno' in product_lower or 'collagen' in product_lower or 'colageno' in product_lower:
                translated_products.append('Collagen')
            elif 'vitamina' in product_lower or 'vitamin' in product_lower:
                # Identificar tipos especficos de vitaminas
                if 'vitamina e' in product_lower or 'vitamin e' in product_lower:
                    translated_products.append('Vitamin E')
                elif 'vitamina c' in product_lower or 'vitamin c' in product_lower:
                    translated_products.append('Vitamin C')
                elif 'vitamina d' in product_lower or 'vitamin d' in product_lower:
                    translated_products.append('Vitamin D')
                elif 'vitamina b' in product_lower or 'vitamin b' in product_lower:
                    translated_products.append('Vitamin B Complex')
                else:
                    translated_products.append('Multivitamins')
            elif 'omega' in product_lower:
                translated_products.append('Omega 3/6/9')
            elif 'antioxidante' in product_lower or 'antioxidant' in product_lower:
                translated_products.append('Antioxidants')
            elif 'mineral' in product_lower or 'minerals' in product_lower:
                translated_products.append('Minerals')
            elif 'probiotico' in product_lower or 'probiotic' in product_lower:
                translated_products.append('Probiotics')
            elif 'coenzima' in product_lower or 'coenzyme' in product_lower or 'q10' in product_lower:
                translated_products.append('Coenzyme Q10')
            elif 'cido hialurnico' in product_lower or 'hyaluronic' in product_lower:
                translated_products.append('Hyaluronic Acid')
            elif 'biocel' in product_lower or 'biocell' in product_lower:
                translated_products.append('Biocell Collagen')
            elif 'resveratrol' in product_lower:
                translated_products.append('Resveratrol')
            elif 'curcuma' in product_lower or 'turmeric' in product_lower:
                translated_products.append('Turmeric/Curcumin')
            elif 'verisol' in product_lower:
                translated_products.append('Verisol Collagen')
            elif 'astaxantina' in product_lower or 'astaxanthin' in product_lower:
                translated_products.append('Astaxanthin')
            else:
                translated_products.append(product.title())
        all_products.extend(translated_products)

product_counts = pd.Series(all_products).value_counts()

# Crear grfica de barras horizontales
colors = oranges_palette[:len(product_counts)]
bars = plt.barh(range(len(product_counts)), product_counts.values, color=colors,
                alpha=0.9, edgecolor='white', linewidth=1.5)
plt.yticks(range(len(product_counts)), product_counts.index)
plt.xlabel('Number of Respondents', fontweight='bold')
plt.ylabel('Type of Nutricosmetic', fontweight='bold')
plt.title('Specific Nutricosmetics Used by Respondents', fontsize=16, fontweight='bold')
plt.grid(True, alpha=0.3, axis='x')

# Aadir etiquetas de valor en las barras
for i, v in enumerate(product_counts.values):
    plt.text(v + 0.5, i, str(v), ha='left', va='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

# 3. GRFICA: Razones de compra por grupo de edad (en ingls)
plt.figure(figsize=(12, 8))
purchase_reasons = df[reasons_col].dropna()
age_groups = df[age_col].dropna()

min_length = min(len(purchase_reasons), len(age_groups))
reason_age_df = pd.DataFrame({
    'Purchase_Reason': purchase_reasons.iloc[:min_length],
    'Age_Group': age_groups.iloc[:min_length]
})

reason_age_count = reason_age_df.groupby(['Age_Group', 'Purchase_Reason']).size().unstack(fill_value=0)
top_reasons = reason_age_df['Purchase_Reason'].value_counts().head(5).index
reason_age_filtered = reason_age_df[reason_age_df['Purchase_Reason'].isin(top_reasons)]
reason_age_count_filtered = reason_age_filtered.groupby(['Age_Group', 'Purchase_Reason']).size().unstack(fill_value=0)

# Normalizar para vista porcentual
reason_age_percentage = reason_age_count_filtered.div(reason_age_count_filtered.sum(axis=1), axis=0) * 100

reason_age_percentage.plot(kind='bar', width=0.85, color=oranges_palette[:5],
                          edgecolor='white', linewidth=1.5, alpha=0.9)
plt.xlabel('Age Group', fontweight='bold')
plt.ylabel('Percentage of Respondents (%)', fontweight='bold')
plt.title('Top 5 Purchase Reasons by Age Group', fontsize=16, fontweight='bold')
plt.legend(title='Purchase Reasons', bbox_to_anchor=(1.05, 1), loc='upper left',
           frameon=True, fancybox=True, shadow=True)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 4. GRFICA: Distribucin por edad (en ingls)
plt.figure(figsize=(10, 8))
age_distribution = df[age_col].value_counts().sort_index()

# Asegurar que los rangos de edad estn en orden lgico
age_order = ['Under 18', '18-25', '26-35', '36-45', '46-55', '56-65', '65+']
age_distribution = age_distribution.reindex([age for age in age_order if age in age_distribution.index])

colors = oranges_palette[:len(age_distribution)]
wedges, texts, autotexts = plt.pie(age_distribution.values,
                                   labels=age_distribution.index,
                                   autopct='%1.1f%%',
                                   startangle=90,
                                   colors=colors,
                                   wedgeprops={'edgecolor': 'white', 'linewidth': 1.5})
plt.title('Age Group Distribution', fontsize=16, fontweight='bold')

# Mejorar apariencia del texto
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(11)

plt.tight_layout()
plt.show()

# 5. GRFICA: Uso de nutricosmticos por grupo de edad
plt.figure(figsize=(12, 8))
usage_by_age = pd.crosstab(df[age_col], df[used_col])
usage_percentage = usage_by_age.div(usage_by_age.sum(axis=1), axis=0) * 100

# Reordenar por edad si es necesario
usage_percentage = usage_percentage.reindex(age_order, errors='ignore')

usage_percentage.plot(kind='bar', color=[oranges_palette[2], oranges_palette[6]],
                     edgecolor='white', linewidth=1.5, alpha=0.9)
plt.xlabel('Age Group', fontweight='bold')
plt.ylabel('Percentage (%)', fontweight='bold')
plt.title('Nutricosmetic Usage by Age Group', fontsize=16, fontweight='bold')
plt.legend(title='Has Used', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# Resumen estadstico
print("="*70)
print("DESCRIPTIVE STATISTICS: NUTRICOSMETICS CONSUMER STUDY")
print("="*70)
print(f"Total respondents: {len(df)}")
print(f"Have heard about nutricosmetics: {heard_counts[0]} ({heard_counts[0]/len(df)*100:.1f}%)")
print(f"Have used nutricosmetics: {used_counts[0]} ({used_counts[0]/len(df)*100:.1f}%)")
print(f"Most common age group: {age_distribution.idxmax()} ({age_distribution.max()/len(df)*100:.1f}%)")
print(f"Most used nutricosmetic: {product_counts.index[0]} ({product_counts.values[0]} users)")

if not purchase_reasons.empty:
    top_reason = purchase_reasons.value_counts().idxmax()
    top_reason_count = purchase_reasons.value_counts().max()
    print(f"Most common purchase reason: {top_reason} ({top_reason_count/len(purchase_reasons)*100:.1f}%)")

print("="*70)
print("TOP 5 NUTRICOSMETICS USED:")
for i, (product, count) in enumerate(product_counts.head().items(), 1):
    print(f"{i}. {product}: {count} users ({count/len(products_used)*100:.1f}%)")
print("="*70)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import warnings
warnings.filterwarnings('ignore')

# Configuracin de estilo profesional
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 12
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

# CARGAR LOS DATOS
file_path = '/content/drive/MyDrive/Proyecto Zanahoria/Data_Agrosavia/Estudio de percepcion y perfil del consumidor de Nutricosmeticos - Actualizado1.xlsx'
df = pd.read_excel(file_path)

# IDENTIFICAR COLUMNAS
knowledge_col = df.columns[32]
usage_col = df.columns[35]
motivation_col = df.columns[38]
age_col = df.columns[8]
gender_col = df.columns[11]

# PREPROCESAMIENTO INICIAL - CREAR VARIABLES NECESARIAS PRIMERO
print(" PREPROCESSING DATA")
print("="*50)

# Crear variables binarias primero
df['usage_binary'] = df[usage_col].map({'Si': 1, 'S': 1, 'No': 0}).fillna(0)
df['knowledge_binary'] = df[knowledge_col].map({'Si': 1, 'S': 1, 'No': 0}).fillna(0)

# Crear grupos de edad
df['age_group'] = df[age_col].map({
    'Entre 12 y 18 aos': '12-18',
    'Entre 19 y 26 aos': '19-26',
    'Entre 27 y 59 aos': '27-59',
    'Mayor de 60 aos': '60+',
    'Generacin X (1969 -1980)': 'Gen X (40-54)'
}).fillna('Other')

# Crear variable de gnero
df['gender'] = df[gender_col].map({'Femenino': 'Female', 'Masculino': 'Male'}).fillna('Other')

# Crear variable de motivacin
def extract_motivation(text):
    if pd.isna(text): return 'Other'
    text = str(text).lower()
    if any(word in text for word in ['belleza', 'piel', 'cabello', 'uas', 'apariencia']): return 'Beauty'
    elif any(word in text for word in ['salud', 'enfermedad', 'bienestar', 'cuidado']): return 'Health'
    elif any(word in text for word in ['complemento', 'nutriente', 'suplemento', 'aliment']): return 'Nutrition'
    elif any(word in text for word in ['envejecimiento', 'arrugas', 'anti-age']): return 'Anti-aging'
    elif any(word in text for word in ['energa', 'cansancio', 'fatiga']): return 'Energy'
    elif any(word in text for word in ['deporte', 'ejercicio', 'gimnasio']): return 'Sports'
    else: return 'Other'

df['purchase_motivation'] = df[motivation_col].apply(extract_motivation)

# 1. ANLISIS ESTADSTICO CON TUKEY - AHORA S FUNCIONAR
print("\n STATISTICAL ANALYSIS WITH TUKEY TEST")
print("="*50)

# ANOVA para conocimiento por grupos de edad
print("1. ANOVA - Knowledge by Age Groups:")
try:
    model_knowledge = ols('knowledge_binary ~ C(age_group)', data=df).fit()
    anova_knowledge = sm.stats.anova_lm(model_knowledge, typ=2)
    print(anova_knowledge)

    if anova_knowledge['PR(>F)'][0] < 0.05:
        print("\n Significant differences found! Performing Tukey HSD test:")
        tukey_knowledge = pairwise_tukeyhsd(endog=df['knowledge_binary'],
                                           groups=df['age_group'], alpha=0.05)
        print(tukey_knowledge)

        # Visualizacin de resultados Tukey
        plt.figure(figsize=(10, 6))
        tukey_knowledge.plot_simultaneous()
        plt.title('Tukey HSD Test - Knowledge by Age Group\n(95% confidence intervals)',
                 fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('Difference in Knowledge Proportion', fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()
    else:
        print("\n No significant differences in knowledge across age groups")

except Exception as e:
    print(f"Error in ANOVA for knowledge: {e}")

# ANOVA para uso por grupos de edad
print("\n2. ANOVA - Usage by Age Groups:")
try:
    model_usage = ols('usage_binary ~ C(age_group)', data=df).fit()
    anova_usage = sm.stats.anova_lm(model_usage, typ=2)
    print(anova_usage)

    if anova_usage['PR(>F)'][0] < 0.05:
        print("\n Significant differences found! Performing Tukey HSD test:")
        tukey_usage = pairwise_tukeyhsd(endog=df['usage_binary'],
                                      groups=df['age_group'], alpha=0.05)
        print(tukey_usage)

        # Visualizacin de resultados Tukey
        plt.figure(figsize=(10, 6))
        tukey_usage.plot_simultaneous()
        plt.title('Tukey HSD Test - Usage by Age Group\n(95% confidence intervals)',
                 fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('Difference in Usage Proportion', fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()
    else:
        print("\n No significant differences in usage across age groups")

except Exception as e:
    print(f"Error in ANOVA for usage: {e}")

# 2. ANLISIS DE CONOCIMIENTO VS USO CON DISEO MEJORADO
print("\n KNOWLEDGE VS USAGE ANALYSIS")
print("="*50)

knowledge_usage = pd.crosstab(df[knowledge_col], df[usage_col])
knowledge_map = {'Si': 'Yes', 'No': 'No', 'S': 'Yes'}
usage_map = {'Si': 'Yes', 'No': 'No', 'S': 'Yes'}

knowledge_usage.index = knowledge_usage.index.map(lambda x: knowledge_map.get(str(x).strip(), str(x).strip()))
knowledge_usage.columns = knowledge_usage.columns.map(lambda x: usage_map.get(str(x).strip(), str(x).strip()))

# Visualizacin mejorada
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

# Heatmap estilizado
im = ax1.imshow(knowledge_usage.values, cmap='RdYlGn', aspect='auto', vmin=0, vmax=knowledge_usage.values.max())
for i in range(len(knowledge_usage.index)):
    for j in range(len(knowledge_usage.columns)):
        ax1.text(j, i, f'{knowledge_usage.values[i, j]}\n({knowledge_usage.values[i, j]/knowledge_usage.sum().sum()*100:.1f}%)',
                ha="center", va="center", color="black", fontsize=14, fontweight='bold',
                bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))

ax1.set_xticks(np.arange(len(knowledge_usage.columns)))
ax1.set_yticks(np.arange(len(knowledge_usage.index)))
ax1.set_xticklabels(knowledge_usage.columns, fontweight='bold')
ax1.set_yticklabels(knowledge_usage.index, fontweight='bold')
ax1.set_xlabel('Usage of Nutricosmetics', fontsize=14, fontweight='bold')
ax1.set_ylabel('Awareness of Nutricosmetics', fontsize=14, fontweight='bold')
ax1.set_title('Knowledge vs Usage Matrix\n(Counts and Percentages)', fontsize=16, fontweight='bold', pad=20)

# Grfico de barras con diseo mejorado
categories = ['Aware & User', 'Aware Non-User', 'Unaware User', 'Unaware Non-User']
values = [
    knowledge_usage.loc['Yes', 'Yes'] if 'Yes' in knowledge_usage.index and 'Yes' in knowledge_usage.columns else 0,
    knowledge_usage.loc['Yes', 'No'] if 'Yes' in knowledge_usage.index and 'No' in knowledge_usage.columns else 0,
    knowledge_usage.loc['No', 'Yes'] if 'No' in knowledge_usage.index and 'Yes' in knowledge_usage.columns else 0,
    knowledge_usage.loc['No', 'No'] if 'No' in knowledge_usage.index and 'No' in knowledge_usage.columns else 0
]

colors = ['#2ecc71', '#f39c12', '#e74c3c', '#95a5a6']
bars = ax2.bar(categories, values, color=colors, alpha=0.9, edgecolor='black', linewidth=1.5)

for bar, value in zip(bars, values):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.5,
            f'{value}\n({value/sum(values)*100:.1f}%)' if sum(values) > 0 else f'{value}\n(0.0%)',
            ha='center', va='bottom', fontweight='bold', fontsize=12,
            bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))

ax2.set_ylabel('Number of Respondents', fontsize=14, fontweight='bold')
ax2.set_title('Consumer Awareness and Usage Patterns', fontsize=16, fontweight='bold', pad=20)
ax2.tick_params(axis='x', rotation=45)
ax2.grid(axis='y', alpha=0.4, linestyle='--')
ax2.set_axisbelow(True)

plt.tight_layout()
plt.show()

# 3. ANLISIS ESTADSTICO ADICIONAL
print("\n ADDITIONAL STATISTICAL ANALYSIS")
print("="*50)

# Chi-square test para motivacin por gnero
motivation_by_gender = pd.crosstab(df['gender'], df['purchase_motivation'])
print("Chi-square Test - Motivation vs Gender:")
chi2, p, dof, expected = stats.chi2_contingency(motivation_by_gender)
print(f"Chi2: {chi2:.3f}, p-value: {p:.4f}")
print(" Significant association" if p < 0.05 else " No significant association")

# Chi-square test para motivacin por edad
motivation_by_generation = pd.crosstab(df['age_group'], df['purchase_motivation'])
print("\nChi-square Test - Motivation vs Age:")
chi2_age, p_age, dof_age, expected_age = stats.chi2_contingency(motivation_by_generation)
print(f"Chi2: {chi2_age:.3f}, p-value: {p_age:.4f}")
print(" Significant association" if p_age < 0.05 else " No significant association")

# 4. VISUALIZACIONES MEJORADAS
print("\n PURCHASE MOTIVATION ANALYSIS")
print("="*50)

# Motivacin por generacin
plt.figure(figsize=(16, 10))
colors = sns.color_palette("husl", len(motivation_by_generation.columns))
bars = motivation_by_generation.plot(kind='bar', stacked=True, color=colors, alpha=0.9, edgecolor='black', linewidth=1)

plt.title('Purchase Motivation by Generation\nConsumer Behavior Analysis', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Age Group', fontsize=16, fontweight='bold')
plt.ylabel('Number of Respondents', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontweight='bold')
plt.legend(title='Purchase Motivation', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

# Motivacin por gnero
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
motivation_by_gender.plot(kind='bar', stacked=True, color=colors, alpha=0.9, edgecolor='black', linewidth=1, ax=ax1)
ax1.set_title('Purchase Motivation by Gender\nAbsolute Counts', fontsize=16, fontweight='bold')
ax1.set_xlabel('Gender', fontsize=14, fontweight='bold')
ax1.set_ylabel('Number of Respondents', fontsize=14, fontweight='bold')

motivation_percentage = motivation_by_gender.div(motivation_by_gender.sum(axis=1), axis=0) * 100
motivation_percentage.plot(kind='bar', stacked=True, color=colors, alpha=0.9, edgecolor='black', linewidth=1, ax=ax2)
ax2.set_title('Purchase Motivation by Gender\nPercentage Distribution', fontsize=16, fontweight='bold')
ax2.set_xlabel('Gender', fontsize=14, fontweight='bold')
ax2.set_ylabel('Percentage (%)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# 5. RESUMEN ESTADSTICO
print("\n STATISTICAL SUMMARY")
print("="*50)
print(f"Total respondents: {len(df):,}")
print(f"\nGender distribution:\n{df['gender'].value_counts()}")
print(f"\nAge group distribution:\n{df['age_group'].value_counts()}")
print(f"\nPurchase motivation distribution:\n{df['purchase_motivation'].value_counts()}")

# Tasa de uso por generacin
usage_rate = df.groupby('age_group')['usage_binary'].mean() * 100
print(f"\nUsage rate by generation (%):")
print(usage_rate.round(1))

# Tasa de conocimiento por generacin
knowledge_rate = df.groupby('age_group')['knowledge_binary'].mean() * 100
print(f"\nKnowledge rate by generation (%):")
print(knowledge_rate.round(1))

print("\n" + "="*50)
print("ANALYSIS COMPLETED ")
print("="*50)
```
